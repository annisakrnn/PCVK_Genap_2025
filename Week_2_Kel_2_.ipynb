## import library and image from drive

```python
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
import turtle
img = cv.imread('/content/drive/MyDrive/KELAS 3/PCVK/TUGAS/ktm.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

## 1. Drawing From Rizqi Fauzan

```python
center_x = 220
center_y = 500
star_size = 20

img_with_star = img.copy()

points = []
for i in range(5):
    angle = 2 * np.pi * i / 5 + np.pi / 2
    x = int(center_x + star_size * np.cos(angle))
    y = int(center_y + star_size * np.sin(angle))
    points.append((x, y))

for i in range(5):
    cv.line(img_with_star, points[i], points[(i + 2) % 5], (255, 255, 0), 3)

plt.imshow(img_with_star)
```

## Drawing from Annisa Kurniawati

```python
img_combined = img.copy()
# Gambar bintang
center_x_star = 220
center_y_star = 500
star_size = 20
points = []
for i in range(5):
    angle = 2 * np.pi * i / 5 + np.pi / 2
    x = int(center_x_star + star_size * np.cos(angle))
    y = int(center_y_star + star_size * np.sin(angle))
    points.append((x, y))
for i in range(5):
    cv.line(img_combined, points[i], points[(i + 2) % 5], (255, 255, 0), 3)

# Gambar persegi panjang
center_x_rect = 350
center_y_rect = 500
rect_width = 320
rect_height = 40
top_left = (center_x_rect - rect_width // 2, center_y_rect - rect_height // 2)
bottom_right = (center_x_rect + rect_width // 2, center_y_rect + rect_height // 2)
cv.rectangle(img_combined, top_left, bottom_right, (255, 255, 0), 3)

# Fungsi untuk menggambar matahari
def draw_petal(img, center_x, center_y, size, angle):
    petal_points = []
    for i in range(2):
        rad = np.radians(angle + 180 * i)
        x = int(center_x + size * np.cos(rad))
        y = int(center_y + size * np.sin(rad))
        petal_points.append((x, y))
    cv.line(img, (center_x, center_y), petal_points[0], (0, 255, 0), 3)
    cv.line(img, petal_points[0], petal_points[1], (0, 255, 0), 3)
    cv.line(img, petal_points[1], (center_x, center_y), (0, 255, 0), 3)

center_x_mth = 260
center_y_mth = 500
mth_size = 25

for i in range(8):
    angle = 360 * i / 8
    draw_petal(img_combined, center_x_mth, center_y_mth, mth_size, angle)

cv.circle(img_combined, (center_x_mth, center_y_mth), 15, (255, 255, 0), -1)

plt.imshow(img_combined)
plt.title('Combined Image')
plt.show()
```
