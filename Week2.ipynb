```python
images = {
    "coffee": data.coffee()
}

plt.figure(figsize=(6,6))

for i, (name, img) in enumerate(images.items(), 1):

    if len(img.shape) == 2:
        img_color = cv.cvtColor(img, cv.COLOR_GRAY2BGR)
    else:  # RGB
        img_color = cv.cvtColor(img, cv.COLOR_RGB2BGR)

    cv.putText(img_color,
               text=name,
               org=(20, 40),
               fontFace=cv.FONT_HERSHEY_SIMPLEX,
               fontScale=1.2,
               color=(0, 255, 0),
               lineType=cv.LINE_AA)


    plt.subplot(1, 1, 1)
    plt.imshow(cv.cvtColor(img_color, cv.COLOR_BGR2RGB))
    plt.title(name)
    plt.axis("off")

plt.show()
```
